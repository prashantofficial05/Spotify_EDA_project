```python
##Data Analysis
#Top 5 most popular artists
```

```python
top_artist = data.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:5]
top_artist
```

```python
top_artist.plot.barh()
plt.show()
```

```python
#top 5 loudness tracks
```

```python
top_loudness_track = data[["loudness","song_title"]].sort_values(by="loudness",ascending=True)[:5]
top_loudness_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y="song_title", data=top_loudness_track)
plt.title("Top 5 loudness track ")
plt.show()
```

```python
#Artist with the most danceability song
```

```python
top_artist_danceable_song = data[["danceability","song_title","artist"]].sort_values(by="danceability", ascending=False)[:5]
top_artist_danceable_song 
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x="danceability", y="artist", data=top_artist_danceable_song)
plt.show()
```

```python
#To 10 instrumentalness tracks
```

```python
top_ten_instrumental_track = data[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness", ascending=False)[:5]
top_ten_instrumental_track
```

```python
plt.figure(figsize=(12, 7))
plt.pie(x="instrumentalness", data=top_ten_instrumental_track, autopct='%1.2f%%', labels=top_ten_instrumental_track.song_title)
plt.show()
```

```python
## Multiple feature plot
```

```python
interest_feature_cols = ["tempo","loudness","acousticness","danceability","duration_ms","energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data =data[data["target"] == 1][feature_col]
    neg_data =data[data["target"] == 0][feature_col]
    
    plt.figure(figsize=(12, 7))
    
    sns.distplot(pos_data, bins=30, label="Positive", color="green")
    sns.distplot(neg_data, bins=30, label="Negative", color="red")
    
    plt.legend(loc="upper right")
    plt.title(f'Positve and Negative Plot for {feature_col}')
    plt.show()  
```
